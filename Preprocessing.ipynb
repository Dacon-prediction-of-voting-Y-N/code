```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test_x.csv')
```

```python
train_score = pd.read_csv('../data/train_score.csv')
test_score = pd.read_csv('../data/test_x_score.csv')

train_score = train_score['score']
test_score = test_score['score']
```

```python
train['score'] = train_score
test['score'] = test_score
```

```python
vote = train['voted']
data = pd.concat([train,test],join='inner')
```

```python
data.shape
```

- wr 합계로 퉁
- wf 합계로 퉁

```python
wr_col=[]
for i in range(1,14):
    if len(str(i))!=2:
        i = '0'+str(i)
    wr_col.append('wr_'+str(i))
wr_col
```

```python
data['wr_score'] = np.sum(data[wr_col],axis=1)
```

```python
data['wf_score'] = data['wf_01'] + data['wf_02'] + data['wf_03']
```

```python
wr_col.append('wf_01')
wr_col.append('wf_02')
wr_col.append('wf_03')
data.drop(wr_col,axis=1,inplace=True)
```

```python
data[data['wf_score'] == 3] # 허언증 시키들
```

```python
a_col=[]
for i in range(ord('a'),ord('t')+1):
    a_col.append('Q'+chr(i)+'A')
answer = data[a_col]
```

```python
outlier_idx =[]
for i in range(answer.shape[0]):
    for j in range(1,6):
        if np.sum(answer.values[i]== np.array([j]*answer.shape[1])) == answer.shape[1]:
            outlier_idx.append(i)
```
