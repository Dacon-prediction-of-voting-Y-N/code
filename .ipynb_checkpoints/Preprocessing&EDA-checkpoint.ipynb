```python
train_outlier_idx =[]
for i in range(answer.shape[0]):
    for j in range(1,6):
        if np.sum(answer.values[i]== np.array([j]*answer.shape[1])) == answer.shape[1]:
            train_outlier_idx.append(i)
```

```python
def show_height(ax):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=10)
def all_hue_count(data,feature,hue,row=1,col=2,figsize=(12,8)):
    fig,(ax1,ax2) = plt.subplots(row,col,figsize=figsize)
    sns.countplot(data=data,x = feature,ax=ax1)
    sns.countplot(data=data,x=feature,hue=hue,ax=ax2)
    show_height(ax1)
    show_height(ax2)
    ax1.set(title='{} all_count'.format(feature))
    ax2.set(title='{} hue_count'.format(feature))
```

```python
same = train.iloc[train_outlier_idx]
ax = sns.countplot(same['voted'])
show_height(ax)
```

```python
ax=sns.countplot(same[same['age_group'] != '10s']['voted'])
show_height(ax)
```

```python
from pycaret.classification import *
from sklearn.preprocessing import OneHotEncoder
def ohe_trans(df, col):
    ohe = OneHotEncoder()
    x = ohe.fit_transform(df[col].values.reshape(-1,1)).toarray()
    tmp = []
    for i in range(df[col].unique().size):
        tmp.append(col[0]+str(i))
        
    ohe_df = pd.DataFrame(x, columns=tmp)
    return ohe_df
```

```python
x = ohe_trans(train,'age_group')
train = pd.concat([train.reset_index(drop=True),x],axis=1)
train.drop('age_group',axis=1,inplace=True)
```

```python
x = ohe_trans(test,'age_group')
test = pd.concat([test.reset_index(drop=True),x],axis=1)
test.drop('age_group',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
def ohe_trans(df, col):
    ohe = OneHotEncoder()
    x = ohe.fit_transform(df[col].values.reshape(-1,1)).toarray()
    tmp = []
    for i in range(df[col].unique().size):
        tmp.append(col[0]+str(i))
        
    ohe_df = pd.DataFrame(x, columns=tmp)
    return ohe_df
```

```python
col = ['race','age_group','gender','religion']
for i in col:
    x = ohe_trans(train,i)
    train = pd.concat([train,x],axis=1)
```

```python
col = ['race','age_group','gender','religion']
for i in col:
    x = ohe_trans(test,i)
    test = pd.concat([test,x],axis=1)
```
