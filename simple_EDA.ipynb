```python
import missingno as msno
msno.matrix(train) # missing value check
```

```python
def show_height(ax):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=10)
```

## 1: 투표 한 사람, 2: 투표 안한사람

```python
ax = sns.countplot(train['voted'])
plt.title('Voted')
show_height(ax)
```

- 대체적으로 안한 사람이 많다.

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,8))
sns.countplot(train['gender'],ax=ax1)
sns.countplot(train['gender'],hue=train['voted'],ax=ax2)
show_height(ax1)
show_height(ax2)
```

- 전체적으로 남자가 많고 성별에 따른 차이는 유의미하지 않다

```python
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(12,8))
sns.countplot(train['age_group'],ax=ax1)
sns.countplot(train['age_group'],hue=train['voted'],ax=ax2)
ax1.set(title='all_count')
ax2.set(title='hue_count')
show_height(ax1)
show_height(ax2)
plt.tight_layout()
```

- 10대들은 대부분 안하고 나머지들이 쫌 많이 함
- 느낌상 나이 제한 때문인듯

```python
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(12,8))
sns.countplot(train['religion'],ax=ax1)
sns.countplot(train['religion'],hue=train['voted'],ax=ax2)
ax1.tick_params(labelrotation=30)
ax2.tick_params(labelrotation=30)
ax1.set(title='all_count')
ax2.set(title='hue_count')
show_height(ax1)
show_height(ax2)
plt.tight_layout()
```

- Protestant가 유일하게 많고 나머지는 비슷비슷
- 적은 것들 other로 통합시켜도 ㄱㅊ을듯

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,5))
sns.countplot(train['familysize'],ax=ax1)
show_height(ax1)
sns.boxplot(train['familysize'],ax=ax2)
```

- white가 아닌 애들이 많이 할줄 알았는데 고건 아니네..

```python
# 1 = country_side(시골) , 2 = 교외 , 3= 도시, 0= 무응답
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,5))
sns.countplot(train['urban'],ax=ax1)
sns.countplot(train['urban'],hue=train['voted'],ax=ax2)
show_height(ax1)
show_height(ax2)
ax1.set(title='all_count')
ax2.set(title='hue_count')
```

도시일수록 은근 참여 많이 안하네

```python
def all_hue_count(data,feature,hue,row=1,col=2,figsize=(12,8)):
    fig,(ax1,ax2) = plt.subplots(row,col,figsize=figsize)
    sns.countplot(data=data,x = feature,ax=ax1)
    sns.countplot(data=data,x=feature,hue=hue,ax=ax2)
    show_height(ax1)
    show_height(ax2)
    ax1.set(title='all_count')
    ax2.set(title='hue_count')
```

```python
all_hue_count(train,'education','voted',figsize=(12,5))
```

- 고졸이 가장 많지만 안한다.
- 학력수준이 높을수록 하는 투표를 많이 하는 경향이 있음

```python
all_hue_count(train,'hand','voted')
```

- 얜 별로 유의미하지가 않다.

```python
all_hue_count(train,'married','voted')
```

결혼 해본 경험이 있는 사람들이 투표율이 높다

```python
all_hue_count(train,'engnat','voted')
```
