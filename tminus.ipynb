```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# bar plot 높이 나타내기

```python
def show_height(ax):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=10)
```

## 예시

```python
train = pd.read_csv('C:/Users/YEO JIMIN/Downloads/data-master/data-master/train.csv',index_col=0)
ax = sns.countplot(train['voted'])
plt.title('Voted')
show_height(ax)
```

# 두개 countplot 그릴 때
#### 하난 그 데이터 자체의 countplot
#### 다른 하난 종속변수로 구분된 countplot

```python
def all_hue_count(data,feature,hue,row=1,col=2,figsize=(12,8)):
    fig,(ax1,ax2) = plt.subplots(row,col,figsize=figsize)
    sns.countplot(data=data,x = feature,ax=ax1)
    sns.countplot(data=data,x=feature,hue=hue,ax=ax2)
    show_height(ax1)
    show_height(ax2)
    ax1.set(title='all_count')
    ax2.set(title='hue_count')
    plt.tight_layout()
```

### 사용방법
- data : 데이터 프레임
- feature : 내가 countplot 그릴 변수명 -> str타입으로 적어야됨
- hue: 구분할 변수 (걍 종속변수 넣어) -> str 타입으로 적어야됨
- row : 행
- col: 열
    - default 값이 1행 2열 이니까 바꾸고 싶으면 넣고 아니면 넣지말고
- figsize: 플랏 사이즈 크기 -> (가로,세로) <--- 무조건 괄호로 넣어야되고 그거 헷갈리면 걍 figsize=(8,5) 이런식으로 ㄱ

# 예시
