```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../data/sample_train.csv')
test = pd.read_csv('../data/sample_test.csv')
```

```python
from pycaret.classification import *
```

```python
clf = setup(data = train, target = 'voted')
```

```python
best_5 = compare_models(sort = 'AUC', n_select = 5)
```

```python
blended = blend_models(estimator_list = best_5, fold = 5, method = 'soft') # soft voting
```

```python
pred_holdout = predict_model(blended)
```

```python
final_model = finalize_model(blended)
```

```python
predictions = predict_model(final_model, data = test)
```
