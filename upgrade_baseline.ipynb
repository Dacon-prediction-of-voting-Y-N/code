```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv',index_col=0)
test = pd.read_csv('test_x.csv',index_col=0)
```

```python
pd.set_option('display.max_columns', 100)
train.head(20)
```

```python
import missingno as msno
msno.matrix(train) # missing value check
```

```python
def show_height(ax):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=10)
```

```python
ax = sns.countplot(train['voted'])
plt.title('Voted')
show_height(ax)
```

- 대체적으로 안한 사람이 많다.

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,8))
plt1= sns.countplot(train['gender'],ax=ax1)
plt2=sns.countplot(train['gender'],hue=train['voted'],ax=ax2)
show_height(plt1)
show_height(plt2)
```

- 전체적으로 남자가 많고 성별에 따른 차이는 유의미하지 않다

```python
plt.figure(figsize=(12,8))
ax=sns.countplot(train['age_group'],hue=train['voted'])
show_height(ax)
```

```python
plt.figure(figsize=(12,8))
ax=sns.countplot(train['religion'],hue=train['voted'])
plt.xticks(rotation = 60)
show_height(ax)
```
